```python
import requests
from bs4 import BeautifulSoup
import openpyxl
import datetime
```

```python
workbook = openpyxl.Workbook() # create a new workbook
worksheet = workbook.active    # create a worksheet
worksheet['A1']='職缺名稱'
worksheet['B1']='公司名稱'
worksheet['C1']='地區'
worksheet['D1']='薪資待遇'
worksheet['E1']='經歷需求'
worksheet['F1']='學歷'
worksheet['G1']='連結'
worksheet['H1']='聯絡人'

page=0
while page<52:
    page += 1
    print('**開始爬取第{}頁'.format(page))
    res=requests.get('https://www.1111.com.tw/search/job?ks=%E5%A4%A7%E6%95%B8%E6%93%9A&page='+str(page))
    soup=BeautifulSoup(res.text)
    
    for job in soup.find_all('ul',class_="srh-body__result")[0].find_all('div',class_="item__job-info"):
        
#         print(job.find_all('a')[0].text)
#         print(">",job.find_all('a')[1]['aria-label'])
#         print(">",job.find_all('div')[3].select('i')[0]['aria-label'])
#         print(">",job.find_all('div')[3].select('i')[1]['aria-label'])
#         print(">",job.find_all('div')[3].select('i')[2]['aria-label'])
#         print(">",job.find_all('div')[3].select('i')[3]['aria-label'])
#         print(">",job.find_all('a')[0]['href'])
        #2階爬蟲:
        res2 = requests.get(job.find_all('a')[0]['href'])
        soup2 = BeautifulSoup(res2.text)
#         print(soup2.find_all('div',class_="job-detail-panel--apply-profile-text")[0].span.text)
        
        worksheet.append([job.find_all('a')[0].text,\
                          job.find_all('a')[1]['aria-label'],\
                          job.find_all('div')[3].select('i')[0]['aria-label'],\
                          job.find_all('div')[3].select('i')[1]['aria-label'],\
                          job.find_all('div')[3].select('i')[2]['aria-label'],\
                          job.find_all('div')[3].select('i')[3]['aria-label'],\
                          job.find_all('a')[0]['href'],\
                          soup2.find_all('div',class_="job-detail-panel--apply-profile-text")[0].span.text
                         ])
        
        print('成功輸入一筆資料！')
        
        
workbook.save('職缺清單{}.xlsx'.format(datetime.date.today())) # save the processed workbook
print('職缺清單製作完成')
```
