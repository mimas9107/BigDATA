```python
corpus=[] #空 list準備放 jieba切割出來的容器
maxpage=20
pagecount=1
while 'paging' in res_json_data:
    for post in res_json_data['data']:
        if 'message' in post:
            #print(post['message'] + '\n========\n')
            corpus+=jieba.cut(post['message'])
    # 讀取下一頁：
    res = requests.get(res_json_data['paging']['next'],headers=headers )
    res_json_data=json.loads(res.text)
    pagecount +=1
    # 亂數選取暫停時間.
    time.sleep(random.choice([1,2,3,4,5]))
    # 限制 bot不要抓超過 maxpage頁
    if pagecount >= maxpage : 
        break

print(corpus)
```

```python
dic={}
for word in corpus:
#### mapping
    if word not in dic:
        dic[word]=1
    else:
        dic[word]=dic[word]+1
print(dic)
```

```python
sorted_word=sorted(dic.items(), key=operator.itemgetter(1), reverse = True)

for ele in sorted_word:
    if len(ele[0]) >= 2:
        print(ele[0], ele[1])
```
